```python
import random
import pandas as pd
import numpy as np
```

```python
columns = ['steak_1','steak_2','steak_3','steak_4','steak_5','steak_6','steak_7',
           'steak_8','steak_9','steak_10','steak_11','steak_12','steak_13','steak_14']
```

```python
people_num = 1200
```

```python
turns = pd.DataFrame(0,index = np.arange(people_num), columns = columns)
multi = pd.DataFrame(1,index = np.arange(people_num), columns = columns)
```

```python
jackpot = 3000
column_index = 0
for col in columns:
    column_index += 1
    for i in range(len(turns)):
        num = random.randint(1,100)
        if (column_index == 1):
            steak = 0
        else:
            steak = turns.iloc[i][column_index - 2]
        multiplier = 1
        if (steak < 0):
            steak = -1
        elif (num <=30):
            steak += 100
        elif (num <= 55):
            steak += 50
        elif (num <= 75):
            steak += 200
        elif (num <= 82):
            steak += 350
        elif (num <= 91):
            jackpot += (steak * .5)
            steak =  0
        elif (num <= 94):
            multiplier = 2
        elif (num <= 96):
            steak += 1000
        elif (num <= 98):
            multiplier = 3
        elif (num <= 99):
            steak += jackpot
            jackpot = 3000
        elif (num <= 100):
            jackpot += (steak * .5)
            steak = -1
        turns[col][i] = steak * multi[col][i]
        if (col != 'steak_14'):
            multi.iloc[i,column_index] = multiplier
```
